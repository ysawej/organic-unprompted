```python
import numpy as np
import pandas as pd
import hvplot.pandas  # noqa
import panel as pn

x = np.linspace(-20, 20, 400)
df = pd.DataFrame({'x': x})

mu_slider = pn.widgets.FloatSlider(
    name='Mean', start=-5.0, end=5.0,
    step=0.1, value=0.0, width=200,
)
sigma_slider = pn.widgets.FloatSlider(
    name='Standard-Deviation', start=0.1, end=5.1,
    step=0.1, value=1.0, width=200,
)

def normal_dist(mu, sigma):
    df['y'] = (1/(np.sqrt(2*np.pi*sigma*sigma))) * np.exp(-1*(df['x']-mu)*(df['x']-mu)/ (sigma*sigma))
    return df.hvplot.line(
        x='x', y='y', title='Interactive Normal Dist',
        frame_width=400,
    )

pn.rx(normal_dist)(mu_slider, sigma_slider)
```
